## Homework Assignment

The busbar costs and Capacity Factors (CF) for nuclear and coal electric generating stations are as follows:

| Unit type | Capacity factor | Capital (Cents/KWHr) | Fuel (Cents/KWHr) | O&M (Cents/KWHr) |
|-----------|-----------------|-----------------------|-------------------|------------------|
| Nuclear   | 90%             | 4.5                  | 0.7               | 1.8              |
| Coal      | 70%             | 2.8                  | 2.7               | 1.1              |
 
### Questions

a) Assuming nuclear fuel costs are fixed, what price will coal fuel costs have to rise (as % of current cost) to make nuclear electric generation economically attractive at the stated capacity factors?

b) What must the nuclear capital cost have to reduce (as % of current cost) to make nuclear electric generation economically attractive at the stated capacity factors?

c) Re-answer Part b) but now assume coal unitâ€™s capacity factors rise to be equal to nuclear capacity factors (i.e., 90%). In doing this, assume that the total yearly capital and O&M costs are independent of capacity factors.


```python
# Defining the given data
capacity_factor_nuclear = 0.90  # 90% as fraction
capacity_factor_coal = 0.70  # 70% as fraction
capacity_factor_coal_equal = 0.90  # Assume coal matches nuclear in part c

# Cost components (in cents/kWh)
capital_cost_nuclear = 4.5
fuel_cost_nuclear = 0.7
o_and_m_cost_nuclear = 1.8

capital_cost_coal = 2.8
fuel_cost_coal = 2.7
o_and_m_cost_coal = 1.1

# Function to calculate total cost for a unit type
def calculate_total_cost(capital_cost, fuel_cost, o_and_m_cost):
    return capital_cost + fuel_cost + o_and_m_cost

# a) Find the coal fuel cost required for nuclear to be economically attractive
def calculate_required_coal_fuel_cost(capital_nuclear, fuel_nuclear, o_and_m_nuclear, capital_coal, o_and_m_coal):
    total_nuclear_cost = calculate_total_cost(capital_nuclear, fuel_nuclear, o_and_m_nuclear)
    total_coal_cost_without_fuel = capital_coal + o_and_m_coal
    required_coal_fuel_cost = total_nuclear_cost - total_coal_cost_without_fuel
    return required_coal_fuel_cost

# b) Calculate the required reduction in nuclear capital cost
def calculate_required_nuclear_capital_reduction(capital_nuclear, fuel_nuclear, o_and_m_nuclear, capital_coal, fuel_coal, o_and_m_coal):
    total_coal_cost = calculate_total_cost(capital_coal, fuel_coal, o_and_m_coal)
    required_capital_cost_nuclear = total_coal_cost - (fuel_nuclear + o_and_m_nuclear)
    reduction_percentage = (capital_nuclear - required_capital_cost_nuclear) / capital_nuclear * 100
    return reduction_percentage

# c) Re-answer b) with coal capacity factor = nuclear capacity factor
def calculate_required_nuclear_capital_reduction_equal_cf(capital_nuclear, fuel_nuclear, o_and_m_nuclear, capital_coal, fuel_coal, o_and_m_coal):
    cf_change_factor = capacity_factor_nuclear / capacity_factor_coal
    # remember that you have to divide the cost by the capacity factor since increasing the CF decreases the cost
    total_coal_cost = calculate_total_cost(capital_coal, fuel_coal / cf_change_factor, o_and_m_coal / cf_change_factor)
    required_capital_cost_nuclear = total_coal_cost - (fuel_nuclear + o_and_m_nuclear)
    reduction_percentage = (capital_nuclear - required_capital_cost_nuclear) / capital_nuclear * 100
    return reduction_percentage, total_coal_cost

# Perform calculations
# Part a
required_coal_fuel_cost = calculate_required_coal_fuel_cost(
    capital_cost_nuclear, fuel_cost_nuclear, o_and_m_cost_nuclear,
    capital_cost_coal, o_and_m_cost_coal
)
required_coal_fuel_cost_percentage = (required_coal_fuel_cost / fuel_cost_coal) * 100

# Part b
nuclear_capital_reduction = calculate_required_nuclear_capital_reduction(
    capital_cost_nuclear, fuel_cost_nuclear, o_and_m_cost_nuclear,
    capital_cost_coal, fuel_cost_coal, o_and_m_cost_coal
)

# Part c
nuclear_capital_reduction_equal_cf, new_coal_cost = calculate_required_nuclear_capital_reduction_equal_cf(
    capital_cost_nuclear, fuel_cost_nuclear, o_and_m_cost_nuclear,
    capital_cost_coal, fuel_cost_coal, o_and_m_cost_coal
)

# Display results
print("Part a:")
print(f"Coal fuel cost must rise to {required_coal_fuel_cost:.2f} cents/kWh, which is {required_coal_fuel_cost_percentage:.2f}% of the current cost.")

print("\nPart b:")
print(f"Nuclear capital cost must reduce by {nuclear_capital_reduction:.2f}% to be economically attractive: \n from {capital_cost_nuclear:.2f} to {capital_cost_nuclear * (1 - nuclear_capital_reduction / 100):.2f} cents/kWh.")

print("\nPart c:")
print(f"With coal capacity factor equal to nuclear, nuclear capital cost must reduce by {nuclear_capital_reduction_equal_cf:.2f}%: \n from {capital_cost_nuclear:.2f} to {capital_cost_nuclear * (1 - nuclear_capital_reduction_equal_cf / 100):.2f} cents/kWh. \n Given that the cost of coal is now {new_coal_cost:.2f} cents/kWh.")
```

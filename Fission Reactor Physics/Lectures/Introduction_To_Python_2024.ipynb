## Functions

Note: to specify multiple output parameters, you can separate them with a comma. In this case, the function returns a tuple, which has to be unpacked with multiple variables. Otherwise, you can collect your outputs between square brackets. In this case, the function will return a list.

Some notes:
+ Python does not require for you to declare your variable type a-priori (as in C++), and the variable type can be changed within the code according to your needs. However, this means that there is no control if the user input the wrong variable type (for example, input a string in a variable that is then used in an arithmetic operation).
+ As input parameters for a function, you can use functions. In this case, Python first executes the functions passed as parameters, and then the external one.
+ All variables defined within a function body are seen only by instructions inside the function itself (local variables)

```python
def my_trig_sum(a,b):
    out1 = np.sin(a)+np.cos(b)
    out2 = np.sin(b)+np.cos(a)
    return out1, out2, [out1, out2]

c,d,e = my_trig_sum(2,3)
f = my_trig_sum(2,3)
print(f'c={c}, d={d}, e={e}')
print(f'f={f}')
```

### Lambda Functions

Lambda functions are functions that are defined anonymously, meaning, without a function name. Lambda functions are defined as lambda arguments: expression.

They can have any number of arguments, but only one expression. As you can see from the example below, the function itself has no name, and it returns a function object.

```python
square = lambda x: x ** 2
summ = lambda x,y : x+y

print(square(2))
print(summ(1,2))
type(square)
```

## IF-ELSE Statements

If-Else statements are code constructs that execute blocks of code only if certain conditions, represented by logical expressions, are met.

Once it reaches a True statement, Python will not check the rest of the conditions. A 'if' statement is always required; 'elif' and 'else' statements are optional.

We can insert more complicated conditional statements using logical operators.

We can also use nested if-else statements, but be careful about the indentation: nested if-else statements should be indented deeper compared to the external one.

```python
def thermostat(temp, desired_temp,time):
    if temp < desired_temp - 5 and time == 'night':
        status = 'Heat'
    elif temp > desired_temp + 5 and time == 'night':
        status = 'AC'
    else:
        status = 'OFF'
    return status

status = thermostat(30,18,'night')
print(status)
```

## Iteration

### FOR Loops

a for-loop is a set of instruction that is repeated for every value in a sequence (lists, tuples, arrays, sequential data types)

```python
# Sum of every integer from 1 to 3

n = 0
for i in range(1,4): # for statement
    n = n + i
print(n)

for c in 'black': # Iteration over a string
    print(c)

s = 'black'
for i in range(len(s)):
    print(s[i])

# Sum even indices numbers
s = 0
a = [2,3,1,3,3]
for i in range(0,len(a),2): # start, end, step
    s += a[i] # This is equivalent to s = s + a[i]
print(s)
```

Using for loops we can unpack dictionaries. The 'items' method returns an object containing a list of keys and value pairs in tuple.

```python
# Unpack dictionaries using the for loop

dict_a = {"One":1, "Two":2, "Three":3}
for key, values in dict_a.items():
    print(key,values)
```

We can assign two different looping variables at the same time.

```python
a = ['One','Two','Three']
b = [1,2,3]

for i,j in zip(a,b):
    print(i,j)
```

### WHILE Loops

A while loop is a set of instruction that is repeated as long as the associated logical expression is true. The condition is tested at the beginning of each iteration.

If the logical expression never changes, the while loop continues until the system runs out of memory.

If you want to interrupt the loop (perhaps because a condition has been fulfilled) you can use the keyword 'break', which stops the loop.

The keyword 'continue', instead, skip the remaining code in the current iteration of the loop, continuing to the next element of the looping array

```python
def have_digits(s):
    out = 0
    # loop through the string
    for c in s:
        # check if the character is a digit
        if c.isdigit():
            out = 1
            print(f'Found digit {c}')
            break
    return out

out = have_digits('weare4together')
print(out)
```

```python
# Determine the number of times 8 can be divided by 2 until the result is less than 1

i = 0
n = 8

while n >= 1:
    n /= 2
    i += 1
print(f'n = {n}, i = {i}')
```

In general:
*   Use for loops when the number of iterations to be perfomed is known
*   Use while loops when the number of iterations is not known a priori

## Fission spectrum $\chi(E)$

$\chi(E)dE$ := average number of fission neutrons emitted with energy E in [E : E + dE] per fission neutron

\begin{equation}
\chi(E) = 0.453\exp(-1.036E)\sinh\sqrt{2.29E}
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter definition
a = 0.453
b = 1.036
c = 2.29

# Energy range
E0 = 0 #MeV
EF = 10 #MeV
k = 500
E = np.linspace(E0,EF,k)

# Spectrum evaluation
X = np.linspace(0,0,k)
for i in range(k):
  X[i] = a * np.exp(-b * E[i]) * np.sinh(np.sqrt(c * E[i]))

# Plot result
plt.plot(E,X,'r')
plt.ylabel('X')
plt.xlabel('E [MeV')
```

An alternative way to avoid rewriting the expression to be evaluated every time the independent variable changes is to use the 'lambda function'

Syntax: function_name = lambda free_variable : function_expression

More info: realpython.com/python-lambda

```python
# Using the parameters already defined above

FissionSpectrum = lambda E: a * np.exp(-b * E) * np.sinh(np.sqrt(c * E))
X_Lam = FissionSpectrum(E)

plt.plot(E,X,'r')
plt.plot(E,X_Lam,'b*')
plt.ylabel('X')
plt.xlabel('E [MeV')
plt.grid()
```

```python
EF_New = 50
k_New = 5000

E_New = np.linspace(E0,EF_New,k_New)

X_New = FissionSpectrum(E_New)

plt.plot(E_New,X_New,'b*')
plt.ylabel('X')
plt.xlabel('E [MeV')
plt.grid()
```
